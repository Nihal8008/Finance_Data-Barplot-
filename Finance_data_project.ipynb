```python
# 'How often do you monitor your investment?',
fig, axes = plt.subplots(1, 2, figsize=(15,5))

sns.countplot(ax=axes[0], x=df['Duration'])
axes[0].set_title('Prefer to keep your money')

sns.countplot(ax=axes[1], x=df['Invest_Monitor'])
axes[1].set_title('Invest Monitor')
```

```python
#Reasons for saving
plt.figure(figsize=(12, 7))  #custom size of graph
sns.countplot(x=df['gender'],hue=df['What are your savings objectives?'],linewidth=3,edgecolor='black')
plt.title("Reasons for savings by men and women respectively")
plt.show()
```

```python
# 'Which investment avenue do you mostly invest in?'

sns.countplot(x=df['Investment_Avenues'],hue= df['gender'])
```
